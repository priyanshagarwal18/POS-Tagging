```python
import os
import xml.etree.ElementTree as ET

# output_path = "/home/rutuja/Downloads/output.txt"
# output_f = open(output_path, "w")

dictionary ={}
files_count=0

output_file = open("dict_given.txt", "w")
total=0

for subdir, dir, file in os.walk(r'./Assignment-files/Test-corpus'):
    for file_name in file:
        files_count = files_count+1
        print("running")
        file_path = subdir + os.sep + file_name
        tree = ET.parse(file_path)
        root = tree.getroot()
        for element in root.iter('w'):
            total +=1
            word = element.text.lower().rstrip()
            tag = element.attrib.get('c5')
            tag_list = tag.split('-')
            for tag in tag_list:
                output_file.write("%s_%s\n" % (word, tag))
                
                                
                
        for element in root.iter("mw"):
            total +=1
            if element.text != None:
                word = element.text.lower().rstrip()
            mw_list = ""
            tag = element.attrib.get("c5")
            for words in element.iter("w"):
                mw_list += words.text.lower().rstrip()
            tag = element.attrib.get('c5')
            tag_list = tag.split('-')
            for tag in tag_list:
                output_file.write("%s_%s\n" % (mw_list, tag))
                
        for element in root.iter("c"):
            total+=1
            word = element.text
            tag = element.attrib.get("c5") 
            tag_list = tag.split('-')
            for tag in tag_list:
                output_file.write("%s_%s\n" % (word, tag))
                    
print(total)
```
